```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mpl_toolkits.mplot3d import Axes3D
```

三維分類

```python
X, y = datasets.make_classification(n_samples=100, n_features=5, n_classes=3, n_clusters_per_class=1, random_state=42)

# 切分數據集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model=SVC(kernel='linear', C=1.0)#C是懲罰係數，C越大，懲罰越大，越準確，但越容易過擬合
model.fit(X_train, y_train)
```

```python
# 顯示三維數據分佈
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 將數據點分配到兩個類別，用不同的顏色表示
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', s=50, edgecolors='k')
```

以下例子為:
生成5組座標->
將之歸類(SVM是監督式)->
放入模型->
預測新數據

```python
X = np.array([[1,2],[9,15],[5,2],[3,4],[7,9]])
plt.scatter(X[:,0],X[:,1],c='r',marker='o',s=50)
```

```python
# 標記兩個類別
y = [0,1,0,0,1]
```

```python
# 創建SVM模型
svm_model = SVC(kernel='linear')
svm_model.fit(X, y)
```

```python
# 繪製數據點和決策邊界
plt.figure(figsize=(8, 6))

# 繪製數據點
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=50, edgecolors='k')

# 繪製決策邊界
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# 創建網格來繪製決策邊界
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50),
                     np.linspace(ylim[0], ylim[1], 50))
Z = svm_model.decision_function(np.c_[xx.ravel(), yy.ravel()])

# 將結果放回圖形
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0], alpha=0.5, linestyles=['-'])

# 繪製支援向量
plt.scatter(svm_model.support_vectors_[:, 0], svm_model.support_vectors_[:, 1],
            s=100, facecolors='none', edgecolors='r', linewidths=1.5, label='Support Vectors')

# 設置標題和標籤
plt.title('SVM Decision Boundary and Support Vectors')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# 顯示圖例
plt.legend()

# 顯示圖形
plt.show()
```
